# 队列(queue)
- 是一种操作受限的线性表数据结构。先进先出
- 支持入队操作：enqueue()；出队操作：dequeue()
- 用数组实现的队列叫作顺序队列。这会产生数据搬移的问题，需要使用循环队列解决。
- 用链表实现的队列叫作链式队列。

## 循环队列
- 首位相连的数组的数据结构
- 队列为空的判断条件是: head == tail
- 队列为满的判断条件是: head指针和tail指针关系：$(tail+1)\%n=head$


## 阻塞队列和并发队列
- 阻塞 队列就是入队、出队操作可以阻塞
- 并发队列就是队列的操作多线程安全
- 使用阻塞队列，可以轻松实现“生产者-消费者模型”
